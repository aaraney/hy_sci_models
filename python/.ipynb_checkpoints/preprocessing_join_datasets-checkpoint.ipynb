```python
import pandas as pd
import numpy as np
import geopandas as gpd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Conversions. 1:
FEET_TO_METERS = 0.3048
SQFEET_TO_SQMETERS = 0.092903
CFS_TO_CMS = 0.028316847
SQMI_TO_SQKM = 2.589988110336
FPS_TO_MPS = 0.3048000
```

```python
# SWOT_ADCP id's paired to GE30m (Peirong’s) dataset using a 200 meter naive (0.001801802 degrees) spatial join
# The df has been preprocessed to remove NaN COMID values from Peirong's dataset
joined = pd.read_pickle('../data/joined_ge30m-site_no_swot-COMID_no_NaN_COMID.p')

# Peirong’s dataset based off of Peirong et al., 2019
rivers_30m = gpd.read_file('../data/rivers_ge30m_clipped_to_us_extent.gpkg')

# USGS HYDRoacoustic dataset in support of the Surface Water Oceanographic Topography satellite mission (HYDRoSWOT)
# https://www.sciencebase.gov/catalog/item/57435ae5e4b07e28b660af55
swot_adcp = gpd.read_file('../data/SWOT_ADCP_Dataset_4326.gpkg')
```

```python
# So, there are more site_no's (3696) in joined than there are COMID's (3359)
# meaning that in some places there is a one to many relationship between COMID's
# and site_no's. This makes sense. 

questionable_sites = joined[joined['site_no'].duplicated(keep=False)]
# 13203758 is on a culvert not on the main stream

# comid: 74053424

# COMMENT OUT IF YOU WANT TO RESAVE
# questionable_sites.to_csv('questionable_site_no_joined.csv')
# questionable sites were handled in QGIS and then saved as a gpkg
handled_questionable_sites = gpd.read_file('../data/questionable_site_no_joined.gpkg').drop(columns=['geometry'])
```

```python
handled_questionable_sites
```

```python
# negation of hanlded questionable site COMID's, so the ones that need to be removed
questionable_sites_to_remove = questionable_sites[~questionable_sites['COMID'].isin(handled_questionable_sites['COMID'])]
```

```python
joined = joined.merge(questionable_sites_to_remove, how='outer', indicator=True).query("_merge == 'left_only'")
```

```python
joined.drop(columns=['_merge'], inplace=True)
```

```python
joined = joined.iloc[:, [0,-1]].copy()
```

```python
# joined

# This works
joined_swot_adcp = joined.merge(swot_adcp.drop(columns=['geometry']), how='left', on='site_no')

joined_rivers_30 = joined.merge(rivers_30m.drop(columns=['geometry']), on='COMID', how='left')

all_joined = joined_rivers_30.merge(joined_swot_adcp, on=['COMID', 'site_no'])
```

```python
## ['site_no',
## SI UNITS ##
# 'COMID', 'order', 'area', 'Sin', 'Slp', 'Elev', 'K', 'P','AI', 'LAI',
# 'SND', 'CLY', 'SLT', 'Urb', 'WTD', 'HW', 'DOR', 'QMEAN','Q2', 'width_m', 'width_DHG',

## Imperical UNITS ##
# 'host', 'agency_cd', 'station_nm',
# 'dec_lat_va', 'dec_long_va','coord_datum_cd',
# mi**2 => 'drain_area_va','contrib_drain_area_va',
# ft => 'alt_va',
# 'alt_datum_cd', 'site_tp_cd', 'ad_site_cd', 'site_visit_start_dt', 'site_visit_start_dt_sg',
# 'data_aging_cd', 'q_meas_no',
# discharge measurement date time => 'q_meas_dt', 
# discharge measurement date and time datum => 'q_meas_td',
# discharge value cfs =>'q_va',
# feet => 'stage_va',
# stage delta during measurement feet =>'stage_diff_va',
# duration of change in stage measurement seconds => 'stage_diff_du',
# Mean index velocity measured during the discharge measurement	feet per second => 'mean_idx_vel_va',
# stream cross-section area ft**2 => 'std_sec_area_va',
# Stream velocity ft/s => 'std_sec_vel_va',
# Discharge adjustment from the measured channels to the final discharge value	cubic feet per second => 'q_adj_va',
# base flow code => 'base_flw_cd',
# q quality => 'q_meas_qual_cd',
# was used in records processing => 'q_meas_used_fg',
# discharge measurement channel number =>'q_meas_chan_nu',
# output rounding code of the stream width ft =>'stream_wdth_va',
# Area of the measurement section ft**2 => 'xsec_area_va',
# mean velocity ft/2 => 'mean_vel_va',
# max velocity ft/sec => 'max_vel_va',
# sigma/mu => 'q_coef_var_va',
# Percent of the total discharge actually measured by the profiler => 'flw_meas_fc',
# ft => 'mean_depth_va', 'max_depth_va',
# 'adcp_freq_cd',
# 'q_meas_type_cd', 'q_meth_cd', 'vel_meth_cd', 
# cfs => 'meas_q_va',
# firm/soft =>'chan_stability_cd', 
# material (sand, gravel, silt) => 'chan_mat_cd',
# Code that describes velocity flow distribution within the cross-section over time => 'vel_dstrb_cd',
# Code describes the vertical velocity distribution within the cross-section; => 'vert_vel_dstrb_cd',
# Round((([meas_q_va]/[q_va]) x 100),2) => 'q1_Percent_va',
# 'q1_Outside5Percent',
# Round(([xsecond_area_va] x [mean_vel_va]),2) cubic feet per second => 'q2_xsec_area_X_mean_vel_va',
# Round((([q2_xsecond_area_X_mean_vel_va]/[q_va]) x 100),2) => 'q2_Percent_q_va', 
# 'q2_q_va_OutsideBounds',
# 'q2_Percent_meas_q_va',
# 'q2_meas_q_va_OutsideBounds',
# 'q3_CalcQ',
# 'q3_PercentCalcQ_q_va',
# 'q3_q_va_OutsideBounds',
# 'q3_PercentCalcQ_meas_q_va',
# 'q3_meas_q_va_OutsideBounds',
# 'q4_xsec_area_D_stream_width',
# 'q4_Percent_xsec_area_D_stream_width',
# 'q4_OutsideBounds',
# 'q5_Mean_GT_Max',
# 'q6_mean_vel_GT_max_vel', 
# 'q6_JRatio',
# 'q6_JRatioGTo582',
# 'q6_JRatioLTo836',
# 'q7_max_vel_D_mean_vel',
# 'q7_max_vel_D_mean_vel_LT2',
# 'q8_stream_width_D_mean_depth',
# 'q8_GT2',
# 'q8_LT200',
# 'q9_stream_width_GT100',
# 'q10_mean_vel_GTo5']
# So, site_no, the first column is in both datasets
# Then, COMID

rivers30_coi = ["site_no", "COMID", 'COMID', 'order', 'area', 'Sin', 'Slp', 
 'Elev', 'K', 'P','AI', 'LAI','SND', 'CLY', 'SLT', 'Urb', 'WTD',
 'HW', 'DOR', 'QMEAN','Q2', 'width_m', 'width_DHG']

swot_coi = [ 'drain_area_va','contrib_drain_area_va', # mi**2
 'alt_va', # ft
 'q_meas_dt', # discharge measurement date time
 'q_meas_td', # discharge measurement date and time datum
 'q_va', # discharge value cfs
 'stage_va', # feet
 'stream_wdth_va', # feet
 'std_sec_area_va', # cross-sectional area ft**2
 'std_sec_vel_va', # stream velocity at cross-section ft/s
 'mean_depth_va', # ft
 'max_depth_va', # ft
 'meas_q_va', # cfs
 'chan_stability_cd', # firm/soft
 'chan_mat_cd', # channel material (sand, gravel, silt)
 'xsec_area_va', # area of the measurement section ft**2
 'mean_vel_va', # mean stream velocity during meas ft/s
 'max_vel_va', # max stream velocity during meas ft/s
 ]

# First iteration columns of interest:
cols_of_interest = rivers30_coi + swot_coi
# ["site_no", "COMID", 'COMID', 'order', 'area', 'Sin', 'Slp', 
#  'Elev', 'K', 'P','AI', 'LAI','SND', 'CLY', 'SLT', 'Urb', 'WTD',
#  'HW', 'DOR', 'QMEAN','Q2', 'width_m', 'width_DHG',
 
#  'drain_area_va','contrib_drain_area_va', # mi**2
#  'alt_va', # ft
#  'q_meas_dt', # discharge measurement date time
#  'q_meas_td', # discharge measurement date and time datum
#  'q_va', # discharge value cfs
#  'stage_va', # feet
#  'std_sec_area_va', # cross-sectional area ft**2
#  'std_sec_vel_va', # stream velocity at cross-section ft/s
#  'mean_depth_va', # ft
#  'max_depth_va', # ft
#  'meas_q_va', # cfs
#  'chan_stability_cd', # firm/soft
#  'chan_mat_cd', # channel material (sand, gravel, silt)
#  'xsec_area_va', # area of the measurement section ft**2
#  'mean_vel_va', # mean stream velocity during meas ft/s
#  'max_vel_va', # max stream velocity during meas ft/s
 
# ]

from functools import partial

def scalar_mutate(df, tranformation_map):
    def scale(col, val, x):
        return x[col] * val
    
    scalar_tranformation_map = {k: partial(scale, k, v) for (k,v) in tranformation_map.items()}
    
    ## for some reason this does not work
    # scalar_tranformation_map = {k: lambda x: x[k] * v for (k,v) in tranformation_map.items()}
    return df.assign(**scalar_tranformation_map)

transformations = {
    'drain_area_va': SQMI_TO_SQKM,
    'contrib_drain_area_va': SQMI_TO_SQKM,# mi**2
    'alt_va': FEET_TO_METERS, # ft
    #  'q_meas_dt': , # discharge measurement date time
    #  'q_meas_td':, # discharge measurement date and time datum
    'q_va': CFS_TO_CMS, # discharge value cfs
    'stage_va': FEET_TO_METERS, # feet
    'std_sec_area_va': SQFEET_TO_SQMETERS, # cross-sectional area ft**2
    'std_sec_vel_va': FPS_TO_MPS, # stream velocity at cross-section ft/s
    'mean_depth_va': FEET_TO_METERS, # ft
    'max_depth_va': FEET_TO_METERS, # ft
    'meas_q_va': CFS_TO_CMS, # cfs
    #  'chan_stability_cd':, # firm/soft
    #  'chan_mat_cd':, # channel material (sand, gravel, silt)
    'xsec_area_va': SQFEET_TO_SQMETERS, # area of the measurement section ft**2
    'mean_vel_va': FPS_TO_MPS, # mean stream velocity during meas ft/s
    'max_vel_va': FPS_TO_MPS, # max stream velocity during meas ft/s
    'stream_wdth_va': FEET_TO_METERS,
}


all_joined_transformed = scalar_mutate(all_joined, transformations)
df = all_joined_transformed[cols_of_interest]
```
