```python
import pandas as pd
import numpy as np
from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_pickle('../data/joined_data_subset.p')
```

```python
p = sns.JointGrid(data=df, x="stage_va", y="max_depth_va")
p.plot_joint(sns.scatterplot)
p.plot_marginals(sns.histplot, kde=True)
```

```python
p = sns.JointGrid(
                  y=np.log10(df["width_m"]), 
                  x=df["max_depth_va"]
                 )
p.plot_joint(sns.scatterplot)
p.plot_marginals(sns.histplot, kde=True)
```

```python
p = sns.JointGrid(data=df, x="width_DHG", y="max_depth_va")
p.plot_joint(sns.scatterplot)
p.plot_marginals(sns.histplot, kde=True)
```
